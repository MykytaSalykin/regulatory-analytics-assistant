```python
path = "data/raw/structured/eba_transparency_2025_raq_statistical_annex.xlsx"
df = pd.read_excel(path, sheet_name="Results")

print(df.columns.tolist())
df.head(15)
```

```python
df2 = df.dropna(axis=1, how="all").copy()
print(df2.columns.tolist())
df2.head(10)
```

```python
df2["Autumn-25"].dropna().head(20).tolist()
```

```python
mask_label = df2["Autumn-25"].apply(lambda x: isinstance(x, str))
mask_value = df2["Unnamed: 2"].notna()

print("label rows:", mask_label.sum())
print("value rows:", mask_value.sum())

df2.loc[mask_label, ["Autumn-25"]].head(10)
df2.loc[mask_value, ["Unnamed: 2"]].head(10)
```

```python
df3 = df2.copy()

# Create a column with the current item label
df3["item_label"] = df3["Autumn-25"].where(mask_label)

# Fill labels down so each value row gets the nearest previous label
df3["item_label"] = df3["item_label"].ffill()

# Keep only rows where we have numeric values
df_values = df3[df3["Unnamed: 2"].notna()].copy()

df_values[["item_label", "Unnamed: 2"]].head(20)
print(df_values.shape)
```

```python
df_tidy = pd.DataFrame({
    "exercise": "RAQ_2025_Autumn",
    "period": "TR2025",
    "item_label": df_values["item_label"],
    "value": pd.to_numeric(df_values["Unnamed: 2"], errors="coerce"),
    "source": "eba_transparency_2025_raq_statistical_annex.xlsx"
})

df_tidy.head(20)
print(df_tidy.shape)
print(df_tidy["value"].isna().mean())
```

```python
period_columns = {
    "Autumn-25": "TR2025_AUTUMN",
    "Unnamed: 6": "TR2025_SPRING",
    "Unnamed: 7": "TR2024_AUTUMN",
    "Unnamed: 8": "TR2024_SPRING",
    "Unnamed: 9": "TR2023_AUTUMN",
    "Unnamed: 10": "TR2023_SPRING",
    "Unnamed: 11": "TR2022_AUTUMN",
}

period_columns
```

```python
records = []

for col, period_code in period_columns.items():
    mask_value = df3[col].notna()

    tmp = pd.DataFrame({
        "exercise": "RAQ_2025_Autumn",
        "period": period_code,
        "item_label": df3.loc[mask_value, "item_label"],
        "value": pd.to_numeric(df3.loc[mask_value, col], errors="coerce"),
        "source": "eba_transparency_2025_raq_statistical_annex.xlsx"
    })

    records.append(tmp)

df_long = pd.concat(records, ignore_index=True)

df_long.shape
```

```python
df_long["period"].value_counts()
df_long["item_label"].nunique()
df_long.head(10)
df_long.tail(10)
```

```python
import hashlib

def make_item_code(text: str) -> str:
    return hashlib.sha1(text.encode("utf-8")).hexdigest()[:12]

df_long["item_code"] = df_long["item_label"].apply(make_item_code)

df_long[["item_code", "item_label"]].drop_duplicates().head(10)
```

```python
df_final = df_long[[
    "exercise",
    "period",
    "item_code",
    "item_label",
    "value",
    "source"
]]

df_final.shape
df_final.head()
```
